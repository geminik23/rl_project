```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-10,10,100)
y_pred = x
y_target = np.zeros(x.shape)
error = y_pred - y_target


def huber_loss(x, delta=1.0):
    return np.where(np.abs(x)<delta,1/2*np.square(x), delta * (np.abs(x)-0.5*delta))

mse = 1/2*np.square(error)
mae = np.abs(error)
hb = huber_loss(error, 1)
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(nrows=1, ncols=3,figsize=(24, 6))

ax0.plot(mse)
ax0.set_title('Mean Squared Error (MSE/L2)')

ax1.plot(mae)
ax1.set_title('Mean Absolute Error (MAE/L1)')

ax2.plot(huber_loss(error, 0.2),'k-')
ax2.plot(huber_loss(error, 0.6),'k.')
ax2.plot(huber_loss(error, 1),'k-.')
ax2.legend(['delta=0.2','delta=0.6','delta=1'])
ax2.set_title('Huber Loss')
plt.show()
```
